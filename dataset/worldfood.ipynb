```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as ticker
```

```python
df=pd.read_csv('worldfoodproduction.csv')
# Print First Five Rows
df.head()
```

```python
df.info()
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
entity = df['Entity'].unique()
entity = list(set(entity))
total_entities = len(entity)
print("Tổng số entity:", total_entities)
```

```python
entity
```

```python
years = df['Year'].unique()

if years.size > 0:
    start_year = years.min()
    end_year = years.max()
    print("Starting year:", start_year)
    print("Last year:", end_year)
else:
    print("No data available.")
```

```python
most_recent_year = df['Year'].max()
recent_year_df = df[df['Year'] == most_recent_year]

for column in recent_year_df.columns:
    if column != 'Entity' and column != 'Year':

        top_country = recent_year_df.loc[recent_year_df[column].idxmax(), 'Entity']
        

        print(f"The top producer for {column} is {top_country}.")
```

```python
afghanistan_df = df[df['Entity'] == 'Afghanistan'].dropna(how='all', subset=df.columns[2:])
afghanistan_df
```

```python
afghanistan_yams = afghanistan_df[['Year', 'Yams  Production (tonnes)']]

print("Afghanistan's yam production (1961-2021):")
print(afghanistan_yams)
```

```python
pd.set_option('display.float_format', '{:,.0f}'.format)
print("\nStatistics of yam production:")
print(afghanistan_yams['Yams  Production (tonnes)'].describe())
print("\nHighest output:", afghanistan_yams['Yams  Production (tonnes)'].max(), 
      "ton (year", afghanistan_yams.loc[afghanistan_yams['Yams  Production (tonnes)'].idxmax(), 'Year'], ")")
print("Lowest output:", afghanistan_yams['Yams  Production (tonnes)'].min(), 
      "ton(year", afghanistan_yams.loc[afghanistan_yams['Yams  Production (tonnes)'].idxmin(), 'Year'], ")")
```

```python
fig, ax = plt.subplots(figsize=(12, 6))
ax.plot(afghanistan_yams['Year'], afghanistan_yams['Yams  Production (tonnes)'],
        marker='o', color='blue', label='Yams Production')
ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: f'{int(x):,}'))
ax.set_title('Yams Production in Afghanistan (1961–2021)')
ax.set_xlabel('Year')
ax.set_ylabel('Production (tonnes)')
ax.legend()
plt.xticks(rotation=60)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
afghanistan_data = df[df['Entity'] == 'Afghanistan'][['Year', 'Yams  Production (tonnes)', 'Tomatoes Production (tonnes)']]
correlation = afghanistan_data['Yams  Production (tonnes)'].corr(afghanistan_data['Tomatoes Production (tonnes)'])
print(f"Pearson correlation coefficient between yams and tomatoes products: {correlation:.3f}")
```

```python
afghanistan_data['Yams Normalized'] = (
    afghanistan_data['Yams  Production (tonnes)'] - afghanistan_data['Yams  Production (tonnes)'].min()
) / (afghanistan_data['Yams  Production (tonnes)'].max() - afghanistan_data['Yams  Production (tonnes)'].min())
afghanistan_data['Tomatoes Normalized'] = (
    afghanistan_data['Tomatoes Production (tonnes)'] - afghanistan_data['Tomatoes Production (tonnes)'].min()
) / (afghanistan_data['Tomatoes Production (tonnes)'].max() - afghanistan_data['Tomatoes Production (tonnes)'].min())

fig, ax = plt.subplots(figsize=(12, 6))
ax.plot(afghanistan_data['Year'], afghanistan_data['Yams Normalized'],
        marker='o', color='blue', label='Yams Production (Normalized)')
ax.plot(afghanistan_data['Year'], afghanistan_data['Tomatoes Normalized'],
        marker='o', color='red', label='Tomatoes Production (Normalized)')
ax.set_title('Normalized Yams and Tomatoes Production in Afghanistan (1961-2021)')
ax.set_xlabel('Year')
ax.set_ylabel('Normalized Production')
ax.grid(True)
ax.legend()
plt.xticks(rotation=60)
plt.tight_layout()
plt.show()
```

```python

fig, ax = plt.subplots(figsize=(12, 6))
scatter = ax.scatter(afghanistan_data['Tomatoes Normalized'], afghanistan_data['Yams Normalized'],
                     s=100, color='purple', alpha=0.7, edgecolors='black', label='Data Points')

# Thêm đường xu hướng
z = np.polyfit(afghanistan_data['Tomatoes Normalized'], afghanistan_data['Yams Normalized'], 1)
p = np.poly1d(z)
ax.plot(afghanistan_data['Tomatoes Normalized'], p(afghanistan_data['Tomatoes Normalized']),
        color='black', linestyle='--', label=f'Trend (slope={z[0]:.2f})')

# Thêm nhãn năm cho một số điểm (mỗi 10 năm)
for i, row in afghanistan_data.iloc[::10].iterrows():
    ax.annotate(int(row['Year']), (row['Tomatoes Normalized'], row['Yams Normalized']),
                xytext=(5, 5), textcoords='offset points', fontsize=8)

ax.set_title('Normalized Yams vs Tomatoes Production in Afghanistan (1961-2021)')
ax.set_xlabel('Tomatoes Production (Normalized)')
ax.set_ylabel('Yams Production (Normalized)')
ax.grid(True)
ax.legend()
plt.tight_layout()
plt.show()
```
